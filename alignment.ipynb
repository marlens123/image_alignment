```python
# convert .png to .jpg because ossp works only with .jpg
reg_path = "E:/test/alignment/registered/"
jpg_path = "E:/test/alignment/registered_jpg/"

for file in os.listdir(reg_path):
    png_img = cv2.imread(os.path.join(reg_path, file)) 
    cv2.imwrite(os.path.join(jpg_path, "{}.jpg".format(file)), png_img, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```

```python
def convert_to_grayscale(im):
    im[im==1] = 0
    im[im==2] = 170
    im[im==3] = 170
    im[im==4] = 255
    return im
```

```python
# convert classified .tifs to .png
class_path = "E:/test/alignment/vis_train/classified/"

for idx, file in enumerate(os.listdir(class_path)):
    if file.endswith('.tif'):
        im = cv2.imread(os.path.join(class_path, file))
        im_vis = convert_to_grayscale(im)

        plt.imshow(im_vis)
        cv2.imwrite("E:/test/alignment/vis_class/{}".format(file), im_vis)
```
